## Transformation Utilities

```python
"""
Common functions used for EKF estimate.
"""
import numpy as np

def skew(w):
    """
    Take the skew symmetric matrix of w (omega), a (3,) vector.
    """
    mat = np.zeros((3,3))
    mat[0,1] = -w[2]
    mat[0,2] = w[1]
    mat[1,0] = w[2]
    mat[1,2] = -w[0]
    mat[2,0] = -w[1]
    mat[2,1] = w[0]
    return mat

def Omega(w):
    """
    Take the Omega operator of a (3,) vector w(omega). 
    See Eq.47 of Trawny
    """
    mat = np.zeros((4,4))
    mat[0:3, 0:3] = -skew(w)
    mat[0:3, -1] = w
    mat[-1, 0:3] = -w.T
    return mat

def test_skew():
    w = np.array([1, -1 , 1])
    w_exp = np.array([[0, -1, -1],[1, 0, -1], [1, 1, 0]])
    w_act = skew(w)
    assert np.array_equal(w_exp, w_act), "array not equal! expected: %s, actual: %s" % (w_exp, w_act)
    
test_skew()
```

![Theta Operator](./reference/Theta.png "Theta Operator")

```python

def Theta(w, dt):
    """
    Theta operator of a (3,) vector w(omega). This is a second order approximation, when assuming the 
    rotation w is small.
    See Eq.101 and Eq.103 of Trawny
    """
    w_norm = np.linalg.norm(w)
    
    if w_norm > 1e-20:
        return np.cos(0.5 * w_norm * dt) * np.eye(4) + 1/w_norm * np.sin(0.5 * w_norm * dt) * Omega(w)
    else:
        return np.eye(4) + 0.5 * dt * Omega(w)

def quat_2_rot(q):
    """
    Implementation for extracting rotation matrix from a quaternion. 
    See Eq.62 of http://mars.cs.umn.edu/tr/reports/Trawny05b.pdf
    :param q: a 4 x 1 column vector
    """
    q_x = skew(q[0:3])
    return (2 * q[3] * q[3] - 1) * np.eye(3) - 2 * q[3] * q_x + 2 * q[0:3] @ q[0:3].T

def quat_2_rot2(q):
    """
    Yet another implementation of quaternion to rotation matrix.
    See Eq.77 of Trawny
    :param q: a 4 x 1 column vector
    """
    q_x = skew(q[0:3])
    return np.eye(3) - 2 * q[3] * q_x + 2 * q_x @ q_x

def normalize_quat(q):
    """
    Normalize a quaternion. 
    """
    if q[3] < 0:
        q *= -1
    return q / np.linalg.norm(q)
```

## EKF Propagate
See (https://docs.openvins.com/propagation.html#disc_prop) for derivation on the discrete-time IMU propgation.

```python
import os.path as path
import pandas as pd

g = np.array([0.0, 0.0, 9.81])

def ekf_pred_discrete(st, a, w, dt):
    """
    Single step for EKF prediction using discrete prediction. 
    :param st: previous step state
    :param a: translational acceleration measured from t to t+1
    :param w: rotational velocity measured from t to t+1
    :param dt: the time elapsed between t and t+1
    """
    
    qt, pt, vt, bgt, bat = st[0:4], st[4:7], st[7:10], st[10:13], st[13:16]
    qt = normalize_quat(qt)
    
    Rt = quat_2_rot2(qt)
    w_hat = w - bgt
    a_hat = a - bat
    qn = Theta(w_hat, dt) @ qt
    
    vn = vt - g * dt + Rt.T @ (a_hat) * dt
    pn = pt + vt * dt - 0.5 * (dt ** 2) * g + 0.5 * Rt.T @ (a_hat) * (dt ** 2)
    
    # TODO: Check if this is correct, as current no bias is assumed initially
    bgn, ban = bgt, bat 
    return np.hstack((qn, pn, vn, bgn, ban))

def ekf_pred_rk4(st, a, w, dt):
    """
    Single step for EKF prediction using RK4(Runge-Kutta)
    :param st: previous step state
    :param a: translational acceleration measured from t to t+1
    :param w: rotational velocity measured from t to t+1
    :param dt: the time elapsed between t and t+1
    """
    qt, pt, vt, bgt, bat = st[0:4], st[4:7], st[7:10], st[10:13], st[13:16]
    Rt = quat_2_rot2(qt)
    w_hat = w - bgt
    a_hat = a - bat
    
    omega = Omega(w_hat)
    
    w_norm = np.linalg.norm(w)
    
    if w_norm > 1e-20:
        dq_dt = (np.cos(0.5 * w_norm * dt) * np.eye(4) + 1/w_norm * np.sin(0.5 * w_norm * dt) * omega) @ qt
        dq_dt2 = (np.cos(0.25 * w_norm * dt) * np.eye(4) + 1/w_norm * np.sin(0.25 * w_norm * dt) * omega) @ qt
    else:
        dq_dt = (np.cos(0.5 * dt * w_norm) * (np.eye(4) + 0.5 * dt * omega)) @ qt
        dq_dt2 = (np.cos(0.25 * dt * w_norm) * (np.eye(4) + 0.25 * dt * omega)) @ qt
    
    dR_dt_transpose = quat_2_rot2(dq_dt).T
    dR_dt2_transpose = quat_2_rot2(dq_dt2).T
    
    # formula: k1 = f(tn, yn)
    k1_v_dot = quat_2_rot2(qt).T @ (a_hat) + g
    k1_p_dot = vt
    
    # formula: k2 = f(tn + dt/2, yn + k1*dt/2)
    k1_v = vt + k1_v_dot * dt / 2
    k2_v_dot = dR_dt2_transpose @ a_hat + g
    k2_p_dot = k1_v
    
    # formula: k3 = f(tn + dt/2, yn + k2*dt/2)
    k2_v = vt + k2_v_dot * dt / 2
    k3_v_dot = dR_dt2_transpose @ a_hat + g
    k3_p_dot = k2_v
    
    # formula: k4 = f(tn+dt, yn + k3*dt)
    k3_v = vt + k3_v_dot * dt
    k4_v_dot = dR_dt_transpose @ a_hat + g
    k4_p_dot = k3_v
    
    # formula: yn + 1 = yn + dt/6 * (k1 + 2*k2 + 2*k3 + k4)
    qn = dq_dt
    qn = normalize_quat(qn)
    
    vn = vt + dt/6 * (k1_v_dot + 2*k2_v_dot + 2*k3_v_dot + k4_v_dot)
    pn = pt + dt/6 * (k1_p_dot + 2*k2_p_dot + 2*k3_p_dot + k4_p_dot)
    
    bgn, ban = bgt, bat
    return np.hstack((qn, pn, vn, bgn, ban))
```

```python
def dead_reckoning(t, s0, at, wt, method="discrete", debug=False):
    """
    Dead reckoning. Only use EKF prediction for the next state of the robot.
    """
    assert method in ["discrete", "rk4"], "Unknown method: %s" % method
    states = np.zeros((t.shape[0] - 1, 16)) # q, p, v, bg, ba
    st = s0 # propagate
    for i in range(0, t.shape[0] - 1):
        dt = t[i + 1] - t[i]
        a, w = at[i, :], wt[i, :]
        sn = np.zeros((1, 16))
        if method is "rk4":
            sn = ekf_pred_rk4(st, a, w, dt)
        elif method is "discrete":
            sn = ekf_pred_discrete(st, a, w, dt)
            
        # sn[0:4] = normalize_quat(sn[0:4])
        
        if debug:
            print("Time: %s; dt: %s; orientation: %s; position: %s; velocity: %s; bias-g: %s; bias-a: %s" \
                  % (t[i+1], dt, sn[0:4], sn[4:7], sn[7:10], sn[10:13], sn[13:16]))
        st = sn
        states[i] = st
        # print("Time: %s; Gyro bias: %s; Accelerometer bias: %s" % (t[i+1], st[10:13], st[13:16]))
    return states

def load_gt(gt_path, delimiter=","):
    pose_gt = pd.read_csv(gt_path, delimiter=delimiter,header=None, skiprows=[0]).values
    p = pose_gt[:, 1:4]
    q = pose_gt[:, 4:]

        
    print("Ground-truth pose size: %s" % (pose_gt.shape, ))
    poses = np.zeros((4, 4, pose_gt.shape[0]))
    for i in range(pose_gt.shape[0]):
        # print("Quaternion: %s" % q[i, :])
        poses[0:3, 0:3, i] = quat_2_rot2(q[i, :])
        poses[0:3, -1, i] = p[i, :]
    
    return poses

def predict_and_plot(imu_path, gt_path, s0, title, show_ori=True, method="discrete", debug=False, delimiter=","):
    """
    :param imu_path: the path for the imu measurement
    :param gt_path: the path for the ground-truth pose
    :param s0: initial state
    :param title: title for the plot
    :param show_ori: whether to show the 2-D orientation
    :param method: method for predicting the IMU pose, one of ["discrete", "rk4"]
    :param debug: whether to print debug message
    :param delimiter: delimiter for reading the GT file, one of [" ", ","]
    """
    # data = data[300:, :]
    poses = None
    poses_gt = None
    
    if imu_path:
        data = pd.read_csv(imu_path, header=None).values
        print("Size of the IMU data: %s" % (data.shape, ))
        print("Example record: %s" % data[0, :])

        states = dead_reckoning(data[:, 0], s0, data[:, 1:4], data[:, 4:], method, debug)

        q, p = states[:, 0:4], states[:, 4:7]
        poses = np.zeros((4, 4, states.shape[0]))
    
        for i in range(states.shape[0]):
            poses[0:3, 0:3, i] = quat_2_rot2(q[i, :])
            poses[0:3, -1, i] = p[i, :]
    
    if gt_path:
        poses_gt = load_gt(gt_path, delimiter)
    visualize_trajectory_2d(poses, poses_gt, show_ori=show_ori, title=title)
```

## EKF Covariance Update

![imu state_jacobian](./reference/imu_state_jacobian.png "IMU State Jacobian")

```python
def Theta_ori_error(w_hat, dt):
    """
    Transition matrix for the error of rotation, \theta_hat.
    :param w_hat:, the estimated rotation from sensor reading and bias, a 3 x 1 column vector
    :param dt: , the delta between timestamps;
    See Eq.166 from Trawny and Eq.171 from Trawny
    """
    w_hat_norm = np.linalg.norm(w_hat)
    skew_w_hat = skew(w_hat)
    
    if w_hat_norm > 1e-20:
        return np.cos(w_hat_norm * dt) * np.eye(3) - np.sin(w_hat_norm * dt) * skew_w_hat / w_hat_norm + \
            (1 - np.cos(w_hat_norm * dt)) * (w_hat @ w_hat.T / (w_hat_norm ** 2))
    else:
        return np.eye(3) - dt * skew_w_hat + dt ** 2 / 2 * skew_w_hat @ skew_w_hat
    
    
def Psi_error(w_hat, dt):
    """
    Transition matrix for error of rotation with respect to gyro bias.
    :param w_hat:, the estimated rotation from sensor reading and bias, a 3 x 1 column vector
    :param dt: , the delta between timestamps;
    See Eq.176 and Eq.180 from Trawny
    """
    # TODO: combine this part as these two quantities are repeatedly calculated
    w_hat_norm = np.linalg.norm(w_hat)
    skew_w_hat = skew(w_hat)
    
    if w_hat_norm > 1e-20:
        return - np.eye(3)*dt + 1/(w_hat_norm ** 2) * (1 - np.cos(w_hat_norm * dt)) * skew_w_hat \
            - 1/(w_hat_norm ** 3) * (w_hat_norm * dt - np.sin(w_hat_norm * dt)) * skew_w_hat @ skew_w_hat
    else:
        return - np.eye(3) * (dt ** 2) / 2 * skew_w_hat - (dt ** 3) / 6 * (skew_w_hat @ skew_w_hat)

def error_state_transition_matrix(R_hat, a_hat, w_hat, dt):
    """
    The error state transition from one state to another state.
    """
    dt2 = dt * dt
    psi_error = Psi_error(w_hat, dt)
    Phi = np.zeros((15,15))
    Phi[0:3, 0:3] = Theta_ori_error(w_hat, dt)
    Phi[0:3, 9:12] = psi_error
    
    Phi[3:6, 0:3] = -0.5 * R_hat.T @ skew(a_hat * dt2) 
    Phi[3:6, 3:6] = np.eye(3)
    Phi[3:6, 6:9] = np.eye(3) * dt
    Phi[3:6, 12:15] =  -0.5 * dt2 * R_hat.T
    
    Phi[6:9, 0:3] = -R_hat.T @ skew(a_hat * dt)
    Phi[6:9, 6:9] = np.eye(3)
    Phi[6:9, 12:15] = - dt * R_hat.T
    
    Phi[9:12, 9:12] = np.eye(3)
    
    Phi[12:15, 12:15] = np.eye(3)

    G = np.zeros((15, 12))
    G[0:3, 0:3] = psi_error
    G[3:6, 3:6] = -0.5 * dt2 * R_hat.T
    G[6:9, 3:6] = dt2 * R_hat.T
    G[9:12, 6:9] = dt * np.eye(3)
    G[12:15, 9:12] = dt * np.eye(3)
    return Phi, G

def propagate_cov(P, Qd, R_hat, a_hat, w_hat, dt):
    """
    Propagate a covariance matrix.
    :param P: the covariance matrix in previous timestamp
    :param Qd: the discrete noise matrix
    :param R_hat: the estimated R_G_to_I from previous timestamp
    :param a_hat: the estimated acceleration
    :param w_hat: the estimated rotation velocity
    :param dt: the estimated duration. 
    """
    Phi, G = error_state_transition_matrix(R_hat, a_hat, w_hat, dt)
    
    return Phi @ P @ Phi.T + G @ Qd @ G.T
    
```
